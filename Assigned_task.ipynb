# Combine

```python
tempdf=df1

dff2 = tempdf.groupby(["Offer", "Category", "Brand", "MRP", "Padded"])

tempdf=dff2.agg({'Click Cnt':'mean', 'Sale Qty':'mean','Live Stock':'mean','Initital Stock':'mean','Size Count':'mean'}).rename(columns={'Click Cnt':'Click Cnt', 'Sale Qty':'Sale Qty','Live Stock':'Live Stock', 'Initital Stock':'Initital Stock','Size Count':'Size Count'})

tempdf['count'] = dff2.size()

tempdf["rank"] = tempdf["count"].rank()
tempdf.sort_values("count", inplace = True)

tempdf=tempdf[tempdf['Size Count']>0]

tempdf=tempdf[tempdf['Live Stock']>250]

tempdf=tempdf[tempdf['Initital Stock']>0]

tempdf['diff'] = tempdf['Initital Stock'] - (tempdf['Live Stock'] + tempdf['Sale Qty'])

tempdf['Return Percentage']=(tempdf['diff'] / tempdf['Initital Stock']) * 100.0

print(tempdf['Return Percentage'].mean())

tempdf["Final Ranking"] = tempdf["Return Percentage"].rank()
tempdf.sort_values("Return Percentage", inplace = True)

print(tempdf)

tempdf1=tempdf

from IPython.display import HTML

tempdff1 = pd.DataFrame(tempdf1)
# tempdf1 = pd.DataFrame(np.arange(1000), columns=['data'])

tempdff1.to_csv('submission.csv')


# print('DataFrame:\n', tempdff1)

# # default CSV
# csv_data = tempdff1.to_csv()
# print('\nCSV String:\n', csv_data)

def create_download_link(title = "Download CSV file", filename = "data.csv"):  
    html = '<a href={filename}>{title}</a>'
    html = html.format(title=title,filename=filename)
    return HTML(html)

# create a link to download the dataframe which was saved with .to_csv method
create_download_link(filename='submission.csv')
```
